In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:
import pandas as pd
import numpy as np
from random import shuffle

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
```

# Aggegrate Data
The cell below calculate the means, errors, and lower & upper intervals of the above data.

```python
import scipy.stats as st

# Aggregate Statistics
df_agg = pd.DataFrame({'mean':df.mean(axis=1), 'var':df.var(axis=1),'stdev':df.std(axis=1), 'sem':df.sem(axis=1)})
df_agg['error'] = df.sem(axis=1)*st.norm.ppf(0.975)
df_agg['upper interval'] = df_agg['mean'] + df_agg['error']
df_agg['lower interval'] = df_agg['mean'] - df_agg['error']
df_agg = df_agg.T
```

# Plot Data On Interactive Chart
The data are plotted in a bar chart. A horizontal bar is added and the color of the bars are based on the value of the horizontal line. The colour of each bar is depended on the likelihood that the population mean of the bar is above or below the value of the horizontal bar. Blue bars reprents a population mean below the specified value and red bars represents a population mean above the specified value. The darker the bar colour, the more likely it is to be away from the specified value.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import matplotlib.transforms as transforms
import matplotlib.cm as cm
from matplotlib.colors import Normalize

# import matplotlib.cm as cm
# cmap = cm.autumn
%matplotlib notebook

def t_score(pop_mean, sample_mean, sample_sterr):
    ''' returns the t-score given sample stats'''
    return (sample_mean - pop_mean)/ sample_sterr

def get_bar_prob(y, df):
    '''determine the probability of the population mean
    being smaller/greater than y'''
    bar_prob = []
    df = df.T
    for i in range(df.shape[0]):
        t = t_score(y, df['mean'].iloc[i], df['sem'].iloc[i])
        p = st.norm.cdf(t) - st.norm.cdf(-t)
        bar_prob.append(p)
#         print(f'for bar {i} with mean {df["mean"].iloc[i]}: t-score={t}, prob={p}')   
    return bar_prob

# Setup basic chart
y_val = df_agg.loc['mean'].mean()
fig, ax = plt.subplots()

# Color map
cmap = plt.get_cmap('coolwarm')
norm = Normalize(vmin=-1, vmax=1)
cpick = cm.ScalarMappable(norm=norm, cmap=cmap)
cpick.set_array(np.linspace(-1,1))
fig.colorbar(cpick, orientation='horizontal')

# Plot bars
mybars = ax.bar(df_agg.columns.values, df_agg.loc['mean'], tick_label=df_agg.columns.values, 
       yerr=df_agg.loc['error'].values, capsize=5)
ax.tick_params(axis='x', which='both', bottom=False)
ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda x, p: format(int(x), ',')))

# Change bar colours based on y_val
for bar, prob in zip(mybars, get_bar_prob(y_val, df_agg)):
    bar.set_color(cpick.to_rgba(prob))
fig.canvas.draw_idle()

# Plot horizontal line & show its value
hline_col = '#328B3B'
myhline = ax.axhline(y=y_val, color=hline_col)
mytxt = ax.text(0, y_val,
       f'{y_val:,.0f}', color=hline_col,
        transform=transforms.blended_transform_factory(ax.get_yticklabels()[0].get_transform(), 
                                                       ax.transData),
       ha='right', va='center',
       bbox=dict(facecolor='#FFFFFF', edgecolor=hline_col, alpha=1))
# mytxt = ax.text(1992, y_val, 'test', color=hline_col)

def move_y_value(event):
    '''allows user to interactively change the y value of interest by
    clicking on the graph'''
#     if is_active:
    myhline.set_ydata(event.ydata)
    mytxt.set_position((0, event.ydata))
    mytxt.set_text(f'{event.ydata:,.0f}')
    fig.canvas.draw_idle()
    
def color_bars(event):
    '''changes the relatively bar colours based on the newly picked
    y value'''
    for bar, prob in zip(mybars, get_bar_prob(event.ydata, df_agg)):
        bar.set_color(cpick.to_rgba(prob))
    fig.canvas.draw_idle()


plt.gcf().canvas.mpl_connect('button_press_event', move_y_value)
plt.gcf().canvas.mpl_connect('button_press_event', color_bars)

plt.savefig('ComparisonOfMeans.png')
```
